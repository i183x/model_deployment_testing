```python
import re
import nltk
from nltk.tokenize import word_tokenize
import json

# Read text from JSON file
with open('route_unique.json', 'r') as file:
    data = json.load(file)
    routes = data  # Assuming the JSON file contains a list of strings directly

# Define a regular expression pattern to match "from [origin] to [destination]"
pattern = r"(\w+) to (\w+)"

# Iterate through the routes
for route in routes:
    # Find matches using the regular expression pattern
    matches = re.findall(pattern, route)

    # Extract origin and destination from the matches
    for match in matches:
        origin = match[0]
        destination = match[1]
        print(f"Origin: {origin}, Destination: {destination}")
```

```python
import json

# Read the content of route.txt
with open('route.txt', 'r') as file:
    routes = file.readlines()

# Initialize empty lists for origins and destinations
origins = []
destinations = []

# Parse routes and populate lists
for route in routes:
    origin, destination = route.strip().split(', ')
    origin_city = origin.split(': ')[1]
    destination_city = destination.split(': ')[1]

    # Add to origins list
    origins.append(origin_city)

    # Add to destinations list
    destinations.append(destination_city)

# Write to origin.json
with open('origin.json', 'w') as origin_file:
    json.dump(origins, origin_file, indent=2)

# Write to destination.json
with open('destination.json', 'w') as destination_file:
    json.dump(destinations, destination_file, indent=2)
```

```python
import json

# Read the content of route.txt
with open('route.txt', 'r') as file:
    routes = file.readlines()

# Initialize empty sets for origins and destinations
origins_set = set()
destinations_set = set()

# Parse routes and populate sets
for route in routes:
    origin, destination = route.strip().split(', ')
    origin_city = origin.split(': ')[1]
    destination_city = destination.split(': ')[1]

    # Add to origins set
    origins_set.add(origin_city)

    # Add to destinations set
    destinations_set.add(destination_city)

# Convert sets to sorted lists
origins = sorted(list(origins_set))
destinations = sorted(list(destinations_set))

# Write to origin.json
with open('origin.json', 'w') as origin_file:
    json.dump(origins, origin_file, indent=2)

# Write to destination.json
with open('destination.json', 'w') as destination_file:
    json.dump(destinations, destination_file, indent=2)
```
